esta pacote sserve para operações matriciais e estatisticas básicas.

```python
#Importando numpy
import numpy as np #Simplifica a chamada numpy se quiser colocar 'import numpy', tambem funciona mas sempre tera q escreve 'numpy'
```

```python
c = np.random.rand(100)  #Cada ponto é uma classe dentro de outra classer, (rand é dentro da classe do randon)
print(c)

#Acessando o ultimo valor de 'c'
print('O ultimo valor de c é : ' + str(c[-1]))

```

```python
#Acessando os 10 primeiros valores
print(c[0:10]) #Ele nao pega o ultimo valor '10'


#Comprimento da matriz
print(np.shape(c)) #é uma vetor(100 linhas)
```

```python
#Criando uma matriz numpy
d= np.random.rand(100,100)
print(d)
print(np.shape(d))

print(d[10,10]) #11º valor da 11º linha
```

```python
#Gerando uma matriz 3d

mat3d = np.random.rand(365,100,100)
print(np.shape(mat3d))

```

```python
#Acessando todos os valores da primeira dimensao na celula 49, 49

timeseries = mat3d[:10,49,49]
print(timeseries)
```

```python
timeseries2 = mat3d[:,50:52,50]
print(timeseries2.shape)
```

```python
#Estatisticas com numpy

mediats = np.mean(timeseries2)
print(mediats)

#media das linhas
mediatscolumns = np.mean(timeseries2,axis = 0)  #pega o eixo das lliha e faz as medias
print(mediatscolumns)

```

```python
#media minimo, max,max,percentil, bloplox

#media
media3d = np.mean(mat3d)
print(media3d)

#min
min3d = np.min(mat3d)
print(min3d)

#max
max3d = np.max(mat3d)
print(max3d)

#desvio
std3d = np.std(mat3d)
print(std3d)

#porcentil
pct3d = np.percentile(mat3d,90) # a partir do valor dado, 90% maiores
print(pct3d)

#mediana
mediana3d = np.median(mat3d)
print(mediana3d)

#variança
var3d = np.sqrt(std3d)
print(var3d)

#variança
var3d = np.sqrt(std3d)
print(var3d)


print(media3d.shape)
```

```python
#Concatenate de lista
mat3d2 = mat3d.copy()

matconcat = np.concatenate([mat3d,mat3d2])
print(matconcat.shape)  #Impilha todas os calores de cada eixo
```

# Plotando com matplotlib

* Plot
* Histogram
* bar
* boxplot
* pcolor
* scatter

```python
import matplotlib.pyplot as plt #Importa apenas a classe pyplot do maatplot

#Criando APENAS 1 plot para postar os dados
fig,ax = plt.subplots()  #Deve ser nessa ordem pois, primeiro importa a fig depois o ax
#Lineplo
ax.plot(mat3d[:,10,10])
```

```python


#Criando um plot para postar os dados
fig,ax = plt.subplots(3)  #Deve ser nessa ordem pois, primeiro importa a fig depois o ax
#Lineplo
ax[0].plot(mat3d[:,10,10])
ax[1].plot(mat3d[:,10:12,10])

#bar
ax[2].bar(np.array(range(0,mat3d.shape[0])),mat3d[:,10,10])
```

```python
#Boxplot

fig,ax = plt.subplots(2)
ax[0].boxplot(mat3d[:,10:20,0].flatten()) #Pega todos os valores dess eintervalo e faz tudo em um box
ax[1].boxplot(mat3d[:,10:20,0]) #Box de todoss os valores
```

```python
#Pcolor

fig,ax = plt.subplots(2,2)

ax[0,0].pcolor(mat3d[0,:,:] , cmap = 'jet')
ax[0,1].contour(np.sin(mat3d[0,:,:] ))
ax[1,0].histogram(np.sin(mat3d[0,:,:]).flatten())
```
